```python
import pandas as pd
import os
path='/content/test-sample'
p=[]
cnt=[]
for root, dirs, files in os.walk(path):
  for file in files:
    abs_file = os.path.join(root, file)
    p.append(abs_file)
    try:
      f=open(abs_file, encoding='utf-8', mode='r').read()
      cnt.append(f)
    except UnicodeDecodeError:
      print(file)
      os.remove(abs_file)
```

```python
df1 = pd.DataFrame(p, columns=['path'])
df2= pd.DataFrame(cnt,columns=['context'])
final_df=pd.concat([df1, df2],axis=1)
final_df.head()
```

```python
final_df.drop_duplicates(inplace=True)
final_df.dropna(inplace=True)
final_df["extension"] = final_df["path"].apply(lambda x: x.split('.')[-1])
```

```python
import json
file_path = "/content/languages.json"
with open(file_path, 'r') as fil:
    lang_table = json.load(fil)
```

```python
final_df["possible_language"] = final_df["extension"].replace({ext: lang for lang, exts in lang_table.items() for ext in exts})
```

```python
final_df.head()
```

```python
final_df = final_df.sample(frac=1).reset_index()
```

```python
final_df
```

```python
final_df.drop(['index'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X = final_df.context
y = final_df.possible_language
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer   #tokenizing it  # unigram, bigram and trigram
token_pattern = r"""([A-Za-z_]\w*\b|[!\#\$%\&\*\+:\-\./<=>\?@\\\^_\|\~]+|[ \t\(\),;\{\}\[\]"'`])"""
vectorizer = TfidfVectorizer(token_pattern=token_pattern, max_features=3000, ngram_range=(1, 3),     
                             norm='l1')

```

```python
from sklearn.model_selection import GridSearchCV
param_grid = dict(clf=[LogisticRegression(), RandomForestClassifier(), MLPClassifier()])
grid_search = GridSearchCV(clf_pipe, param_grid=param_grid, cv=3)
grid_search.fit(X_train, y_train)
```
